```python
from Historic_Crypto import HistoricalData
```

```python
new = HistoricalData('ETH-BTC',86400,'2020-9-10-00-00','2021-12-01-00-00').retrieve_data()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [20, 10]
```

```python
ethusd = HistoricalData('ETH-USD',86400,'2020-9-10-00-00','2021-12-01-00-00').retrieve_data()
```

```python
ethusd.open.plot(grid=True, label="USD exchange rate", legend=True)
ethusd.volume.plot(secondary_y=True, label="volume", legend=True)
```

```python
new.open.plot(grid=True, label="BTC exchange rate", legend=True)
new.volume.plot(secondary_y=True, label="volume", legend=True)
```

```python
dailydiff1 = ethusd.open.pct_change()
dailydiff1 = dailydiff1*dailydiff1
plt.plot(ethusd.index,dailydiff1)
```

```python
dailydiff = new.open.pct_change()
dailydiff = dailydiff*dailydiff
plt.plot(new.index,dailydiff)
```
